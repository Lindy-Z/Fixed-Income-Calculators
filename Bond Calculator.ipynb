```python
from math import e

class Bond:
    
    def __init__(self, couponRate, YTM, Face, NOP, Period, year):
        self.couponRate = couponRate
        self.YTM = YTM
        self.Face = Face
        self.NOP = NOP
        self.Period = Period
        self.year = year
    
    def CF(couponRate, YTM, Face, NOP, Period, year):
        resCF = []
        if Period == 'cont':
            coup = couponRate / NOP * Face
            lastyrPMT = coup + Face
            for i in range(1, year*NOP):
                resCF.append(coup)
            resCF.append(lastyrPMT)
            return resCF
        else:
            coup = couponRate / NOP * Face
            lastyrPMT = coup + Face
            for i in range(1, Period):
                resCF.append(coup)
            resCF.append(lastyrPMT)
            return resCF
        
    def PV(couponRate, YTM, Face, NOP, Period, year):
        CFbond = Bond.CF(couponRate, YTM, Face, NOP, Period, year)
        resPV = []
        prdCount = 1
        prdYld = YTM / NOP
        if Period == 'cont':
            while prdCount <= (year * NOP):
                for j in CFbond:
                    discounted = j / (e ** ((prdCount/NOP)*YTM))
                    resPV.append(discounted)
                    prdCount += 1
            return resPV
        else:
            while prdCount <= Period:
                for j in CFbond:
                    discounted = j / ((1 + prdYld) ** prdCount)
                    resPV.append(discounted)
                    prdCount += 1
            return resPV
    
    def BondPrice(couponRate, YTM, Face, NOP, Period, year):
        PVbond = Bond.PV(couponRate, YTM, Face, NOP, Period, year)
        return sum(PVbond)
        
    def weight(couponRate, YTM, Face, NOP, Period, year):
        PVbond = Bond.PV(couponRate, YTM, Face, NOP, Period, year)
        resWT = []
        for k in PVbond:
            resWT.append(k / sum(PVbond))
        return resWT
    
    def duration(couponRate, YTM, Face, NOP, Period, year):
        WTbond = Bond.weight(couponRate, YTM, Face, NOP, Period, year)
        pmtTime = 1/NOP
        pmtCount = 1
        dura = 0
        while pmtCount <= year * NOP:
            for l in WTbond:
                dura += pmtTime * l
                pmtCount += 1
                pmtTime = (1/NOP) * pmtCount
        return dura
    
    def vex(couponRate, YTM, Face, NOP, Period, year):
        WTbond = Bond.weight(couponRate, YTM, Face, NOP, Period, year)
        CFBond = Bond.CF(couponRate, YTM, Face, NOP, Period, year)
        Price = Bond.BondPrice(couponRate, YTM, Face, NOP, Period, year)
        prod = []
       
        if Period == 'cont':
            prdCount = year * NOP
            prd = []
            resVex = 0
            for num in range(1, prdCount + 1):
                prd.append(num/NOP)
            for i in range(0, prdCount):
                resVex += (prd[i]) ** 2 * WTbond[i]
            return resVex
        else:
            for i in range(1, Period+1):
                prod.append(i)
            Weightage = []
            conv = 0
            y = YTM/NOP
            for k in prod:
                conv += (k*(k+1)*CFBond[k-1])/(1+y)**(k+2)
            conv = conv/Price
            return conv
                
     
    
        
#print(Bond.CF(0.06, 0.05, 1000, 1, 4, 4))
print(Bond.PV(0.06, 0.05, 1000, 1, 4, 4))
print(Bond.BondPrice(0.06, 0.05, 1000, 1, 4, 4))
print(Bond.weight(0.06, 0.05, 1000, 1, 4, 4))
#print(Bond.duration(0.06, 0.05, 1000, 1, 4, 4))
print(Bond.vex(0.06, 0.05, 1000, 1, 4, 4))
```
